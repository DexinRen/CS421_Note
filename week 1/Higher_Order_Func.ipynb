**Lambda Form Function**

We call \ as lambda. and the lambda function are consist of 4 parts:
- Backslash (aka. lambda)
- Parameter list
- Arrow
- Body of function

eg. \x -> x + 1

```haskell
(\x -> x + 1) 41
( \x y -> x + y) 1 2
```

### $\eta$-equivalence

An equivlence:
$$
f\equiv \backslash x -> f x
$$
- proof, assuming f is a function ...
$$
f z \equiv (\backslash x -> f x) z
$$

Examples of equivalent functions:
$$
\begin{array}{rl}
plus\ a\ b &= (+)\ a\ b\\
plus\ a &= (+)\ a\\
plus &= (+)\\
\end{array}
$$

Another examples(equivalent func):
$$
\begin{array}{rl}
inc\ x &= x+1\\
inc &= (+)\ 1\\
inc &= (+1)\\
\end{array}
$$

### Curry and Uncurry

we can use `curry` to make a function that takes two parameters a curring function

```haskell
:t curry
:t tplus
tplus :: (Integer, Integer) -> Integer
tplus (a,b) = a + b
ttplus = curry tplus
:t ttplus
```

### Maping and Folding

```haskell
-- sourse code of map function
-- Map can apply function to list, avoid repettion of code (sum, time, minus, etc.)
-- So `map (+) list == sum list`
map :: (a->b) ->[a]->[b]
map f [] = []
map f (x:xs) = f x : map f xs
```
